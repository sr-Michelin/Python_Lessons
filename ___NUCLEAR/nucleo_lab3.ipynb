```python
import numpy as np
from math import exp
```

```python
z1 = 1
z2 = 6
A = 0.923
T9 = 3*10**-2 #(K)
```

```python
E0 = 0.122*(z1**2*z2**2*A)**(1./3.)*T9**(2/3) #(MeV)
E0
```

```python
dE = 0.2368*(z1**2*z2**2*A)**0.16*T9**0.83 #(MeV)
dE
```

```python
kT = 2.5875 * 10**-3 #(MeV)
```

```python
E = np.linspace(10**-30, 2*E0, 1000)
b = 2*E0**1.5/kT
b
```

```python
def y(x):
    return 10**19*exp(-b*x**-0.5 - x/kT)
```

```python
Y = np.vectorize(y)
Y(E)
```

```python
%pylab inline
plot(E*1000, Y(E))
xlabel('E, keV')
ylabel('Y(E)')
```
