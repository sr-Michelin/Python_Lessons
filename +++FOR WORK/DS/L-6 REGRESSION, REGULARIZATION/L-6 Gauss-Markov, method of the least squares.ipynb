# Задачі регресії та регуляризації (класифікація)
Регресія - прогноз певного числа у задачі із вчителем.


### Теорема Гаусова-Маркова

* Ознака y лінійно залежить від від X: $$y = w^TX + error$$.
* Координати даних X(в-р), y(скаляр) - незалежні.
* Щоб врахувати w0, матрицю X доповнюють константною колонкою із одиниць. $$rank X = d+1$$.
* Математичне очікування "errors" при фіксованому "X" рівне нулеві (незміщення). 
* Гетероскедастичність

Коли усі властивості т-ми Гаусова-Маркова виконуються, справджується метод найменших квадратів для максимальної правдоподібності.

### Гетероскедастичність
Гетероскедастичність - зростання розкиду даних у коррдинатній залежності. Дисперсія випадкових помилок при фіксованому X рівна певній константі.

* Зображення з випадковими даними, що показує гетероскедастичність

### Принцип максимальної правдоподібності (likelihood)

Досягнення мінімальної дисперсії для оцінок ваг "w" у класі всіх незміщенних е-лтів X --> задача оптимізації.


### NLL - негативна правдоподібність


## Регуляризація лінійних моделей
У реальних моделях часто виникає проблема мультиколінеарності.
Навчання моделі МН - правильний підхід регуляризації.

Суть регуляризації - обмеження нормованих ваг w певною константою:

* великі ваги моделі можуть викликати явище перенавчання (малі зміни ознаків породжують страшні відхилення)


* Підбір λ* - задача регуляризатора

### Типи регуляризаторів

* Ridge - регресія більш доцільна у проблемах мультиколінеарності
* Lasso (least absolute shrinkage and selection operator) - метод обирає ознаки (features)
